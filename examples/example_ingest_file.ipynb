```python
import os
import boto3
import boto3.s3.transfer


file_path = 'my_file.csv'
with open(file_path, 'w') as f:
    f.write('hello, world')


# =======================
# Upload file using boto3
# =======================

s3 = boto3.client('s3',
    aws_access_key_id=os.environ["AWS_ACCESS_KEY_ID"],
    aws_secret_access_key=os.environ["AWS_SECRET_ACCESS_KEY"],
    endpoint_url=os.environ["AWS_ENDPOINT_URL"],
    region_name=os.environ["AWS_REGION"],
)

s3_path = f'Stage/my_organization/my_file.csv'
bucket = os.environ["TNGRI_S3_BUCKET_NAME"]

config = Config=boto3.s3.transfer.TransferConfig(use_threads=False)
s3.upload_file(file_path, bucket, s3_path, Config=config)

res = s3.list_objects(Bucket=bucket, Prefix=s3_path)
res['Contents']
```

```python
import os 
from tngri.client import Client
from tngri.config import Config


# ===========================================
# Upload file using tngri client (standalone)
# ===========================================

file_path = 'my_file.csv'
with open(file_path, 'w') as f:
    f.write('hello, world')


config = Config.from_env()
config.s3_region=os.environ["TNGRI_S3_REGION"]
config.s3_default_region=os.environ["TNGRI_S3_REGION"]
config.s3_access_key_id=os.environ["AWS_ACCESS_KEY_ID"]
config.s3_secret_access_key=os.environ["AWS_SECRET_ACCESS_KEY"]
config.s3_endpoint_url=os.environ["AWS_ENDPOINT_URL"]
config.s3_bucket_name=os.environ["TNGRI_S3_BUCKET_NAME"]


client = Client(config)
client.upload_file(file_path)
```
