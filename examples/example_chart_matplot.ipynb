```python
create table if not exists cbrf_money_columns as 
select *
from read_json('test.json');


select cast(date as date) as date, 
    max(case when element_id = 12 then obs_val end) as "всего",
    max(case when element_id = 23 then obs_val end) as "денежный агрегат М2",
    max(case when element_id = 24 then obs_val end) as "депозиты в иностранной валюте других финансовых организаций",
    max(case when element_id = 25 then obs_val end) as "депозиты в иностранной валюте нефинансовых организаций",
    max(case when element_id = 26 then obs_val end) as "депозиты в иностранной валюте домашних хозяйств",
    max(case when element_id = 27 then obs_val end) as "депозитные и сберегательные сертификаты"
from cbrf_money_columns
group by 1 
order by 1
```
